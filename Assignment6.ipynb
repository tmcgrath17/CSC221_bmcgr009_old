```python
#Try it yourself

import random
import time
summation = 0
maxi = random.randint(999999,9999999)
count = 0
while summation < maxi:
  num = random.randint(1,999999)
  count += 1
  print(f"\n{num}")
  digits = len(str(num))
  print(f'This number is {digits} digits long')
  if digits <= 4:
    print("Wow, that's pretty low!")
  summation += num
  print(f"The sum is now {summation}")
  time.sleep(1)

print(f"You reached the maximum of {maxi}! It took {count} additions.")
```
